```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.drodd.com/familyguyfun/scripts.html"
response = requests.get(url)
page = BeautifulSoup(response.content, "html5lib")
```

```python
div = page.find(class_="family-guy")
tables = div.find_all("a")
f = open("family-guy.txt", "w+")
len_tables = len(tables)

for i, el_link in enumerate(tables):
    try:
        link = ("https://www.drodd.com/familyguyfun/" + el_link["href"])
        html_script = BeautifulSoup(requests.get(link).content, "html5lib")
        title = "Title: " + html_script.find("h1").text.strip() + "\n\n"
        f.write(title)
        paragraphs =  html_script.find_all("p")
        for p in paragraphs:
            if "gtag" in p.text or "Seth MacFarlane" in p.text:
                print(p.text)
                continue
            f.write(p.text.strip() + "\n")
        f.write("\n\n\n")
        print(f"{i+1} completed out of {len_tables}")
    
    except:
        print("Error")
        continue

f.close()
```
